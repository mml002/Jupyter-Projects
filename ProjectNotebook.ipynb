# YunTea Bobatastic Genie

The YunTea Bobatastic Genie is a Python module that helps users decide what to order from Yun Tea House in San Diego, California. As a chronically indecisive yet adventurous person, especially at boba shops, what better way to try all the menu options at my favorite boba place than to create a Python module that does exactly that? 

Based on their menu options, there are functions that randomly generate drink base recommendations based on milk tea or fruity drink preferences and also toppings too. There are multiple functions that allow users to favorite drinks, keep a blacklist of drinks they don't enjoy, and get descriptions of their drink bases to see whether they'd enjoy the taste. Additionally, users can remove drinks from both the favorites and blacklist lists making it customizable to each user. 

YunTea Bobatastic Genie may help others who also feel indecisive yet adventurous like me wanting to simplify some everyday choices by automating boba orders. Using a variety of Python functions, lists, dictionaries, modules, randomization, and unit testing, this module provides a good example of how coding can be used to explore new things and efficiently manage time while also being able to organize and manage user preferences as well.
